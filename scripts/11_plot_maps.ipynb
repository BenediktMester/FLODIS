# Plot maps
This notebook contains all code to reproduce the map plots used in the paper.

```python
# Import modules
import geopandas as gpd
from shapely import wkt
import pandas as pd
import os
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Create paths

def mkdir(dir):
    
    if not os.path.exists(dir):
        os.mkdir(dir)

path_run = os.getcwd() + '/'
path_data = path_run + 'data/'
path_data_processed = path_run + 'data_processed/'
path_FLODIS_final = path_data_processed + 'FLODIS_final/'
path_world_map = path_data_processed + 'data_world_map/'
path_world_map_creator = mkdir(path_world_map)
path_plots = path_run + 'plots/'
path_plots_creator = mkdir(path_plots)

#GDIS = gpd.read_file(path_data + 'GDIS/GDIS.shp')
countries = gpd.read_file(path_data + 'GADM/gadm36_0.shp')


crs = {'init' :'epsg:4326'} 
```

```python
provinces = gpd.read_file(path_data + 'GADM/gadm36_1.shp')
districts = gpd.read_file(path_data + 'GADM/gadm36_2.shp')
```

# GFD
Load satellite shapefiles and extract centroids

```python
GFD_points = gpd.GeoDataFrame()
GFD_ID_list = []

for GFD_temp_name in sorted(os.listdir(path_data_processed + 'GFD_satellite_shape_files_final/')):
    
    if '.shp' in GFD_temp_name:              
        GFD_temp = gpd.read_file(path_data_processed + 'GFD_satellite_shape_files_final/'+GFD_temp_name)
        GFD_temp_centroid = GFD_temp['geometry'].centroid
        GFD_points = GFD_points.append(GFD_temp_centroid,ignore_index=True)  
        GFD_ID_list.append(GFD_temp_name.split('DFO_')[1].split('_final')[0])

GFD_points['DFO_ID'] = GFD_ID_list
GFD_points.rename(columns={0:'geometry'}, inplace=True) 
GFD_points = gpd.GeoDataFrame(GFD_points,geometry='geometry')
GFD_points.crs = crs  
GFD_points.to_file(path_world_map+'GFD_centroids.shp')
```

# IDMC
Load IDMC geolocation data and extract centroids

```python
# Load all IDMC events
IDMC = pd.read_csv(path_data_processed + 'IDMC_geolocation/IDMC_2008_2021_provinces_districts.csv')
IDMC = IDMC[IDMC.Year <= 2018]
IDMC.rename(columns={'Date of event (start)':'start_date'}, 
                 inplace=True)
IDMC = IDMC[IDMC['Hazard Type']=='Flood']
IDMC.reset_index(drop=True, inplace=True)
IDMC.loc[:,'idx'] = IDMC.index
IDMC_points = gpd.GeoDataFrame()

for IDMC_counter in range(len(IDMC)):
    IDMC_affected_shape = pd.DataFrame()
    IDMC_province_GID_0 = IDMC.ISO3.iloc[IDMC_counter]

    if IDMC.num_provinces.iloc[IDMC_counter] > 0:
        disaster_GID_1s = [x.strip() for x in IDMC.GID_1.iloc[IDMC_counter].split(',')]
        IDMC_affected_shape = IDMC_affected_shape.append(provinces[provinces['GID_1'].isin(disaster_GID_1s)])

    if IDMC.num_districts.iloc[IDMC_counter] > 0:
        disaster_GID_2s = [x.strip() for x in IDMC.GID_2.iloc[IDMC_counter].split(',')]
        IDMC_affected_shape = IDMC_affected_shape.append(districts[districts['GID_2'].isin(disaster_GID_2s)])

    if IDMC_affected_shape.empty == True:
        IDMC_affected_shape = countries[countries['GID_0']==IDMC_province_GID_0] 

    IDMC_temp_centroid = IDMC_affected_shape['geometry'].centroid.reset_index()
    IDMC_points = IDMC_points.append(IDMC_temp_centroid,ignore_index=True)  #['geometry']

IDMC_points = IDMC_points[IDMC_points[0] != 'GeoSeries([], dtype: geometry)']
IDMC_points.rename(columns={0:'geometry'}, inplace=True) 
IDMC_points = gpd.GeoDataFrame(IDMC_points,geometry='geometry')
IDMC_points.crs = crs  
IDMC_points.to_file(path_world_map+'IDMC_centroids.shp')
```

```python
# Load matched IDMC events
#results = pd.read_csv(path_data_processed + 'IDMC_EMDAT_GFD_match/IDMC_GFD_connect_results_FL_merged.csv')
#results = results[results.DFO_matches>0]

results = pd.read_csv(path_FLODIS_final + 'FLODIS_displacements.csv')

matched_ID_list = []

for i in results['Unnamed: 0']:
    matched_ID_list.append(i)
    
IDMC_matched_points = gpd.GeoDataFrame()

for IDMC_counter in range(len(results)):
    IDMC_affected_shape = pd.DataFrame()
    IDMC_province_GID_0 = results.ISO3.iloc[IDMC_counter]

    if results.num_provinces.iloc[IDMC_counter] > 0:
        disaster_GID_1s = [x.strip() for x in results.GID_1.iloc[IDMC_counter].split(',') if type (x) == str]
        IDMC_affected_shape = IDMC_affected_shape.append(provinces[provinces['GID_1'].isin(disaster_GID_1s)])

    if results.num_districts.iloc[IDMC_counter] > 0:
        disaster_GID_2s = [x.strip() for x in results.GID_2.iloc[IDMC_counter].split(',') if type (x) == str]
        IDMC_affected_shape = IDMC_affected_shape.append(districts[districts['GID_2'].isin(disaster_GID_2s)])

    if IDMC_affected_shape.empty == True:
        IDMC_affected_shape = countries[countries['GID_0']==IDMC_province_GID_0] 

    IDMC_temp_centroid = IDMC_affected_shape['geometry'].centroid.reset_index()
    IDMC_matched_points = IDMC_matched_points.append(IDMC_temp_centroid,ignore_index=True) 

IDMC_matched_points = IDMC_matched_points[IDMC_matched_points[0] != 'GeoSeries([], dtype: geometry)']
IDMC_matched_points.rename(columns={0:'geometry'}, inplace=True) 
IDMC_matched_points = gpd.GeoDataFrame(IDMC_matched_points,geometry='geometry')
IDMC_matched_points.crs = crs  
IDMC_matched_points.to_file(path_world_map+'IDMC_centroids_matched.shp')
```

# EM-DAT
Load EM-DAT (GDIS) shapefiles and extract centroids

```python
# Load all EM-DAT events
EMDAT_points = gpd.GeoDataFrame()
EMDAT_ID_list = []

for EMDAT_temp_name in sorted(os.listdir(path_data_processed + 'GDIS_processed/')):
    
    if '.shp' in EMDAT_temp_name:
        EMDAT_temp = gpd.read_file(path_data_processed + 'GDIS_processed/'+EMDAT_temp_name)
        EMDAT_temp_centroid = EMDAT_temp['geometry'].centroid#
        EMDAT_points = EMDAT_points.append(EMDAT_temp_centroid,ignore_index=True)        
        EMDAT_ID_list.append(EMDAT_temp_name.split('mod_')[1].split('.')[0])

EMDAT_points['EMDAT_ID'] = EMDAT_ID_list
EMDAT_points.rename(columns={0:'geometry'}, inplace=True) 
EMDAT_points = gpd.GeoDataFrame(EMDAT_points,geometry='geometry')
EMDAT_points.crs = crs
EMDAT_points.to_file(path_world_map+'EMDAT_centroids.shp')
```

```python
# Load all matched EM-DAT events
results = pd.read_csv(path_FLODIS_final + 'FLODIS_fatalities_damages.csv')
EMDAT_points_matched = gpd.GeoDataFrame()
EMDAT_points_matched['EMDAT_ID'] = results.disasterno
EMDAT_points_matched['geometry'] = np.nan

for EMDAT_disasterno_temp, EMDAT_counter in zip(results.disasterno, range(len(results))):    
    EMDAT_disasterno_temp = [x.strip() for x in EMDAT_disasterno_temp.split(',') if type (x) == str]
    EMDAT_affected_shape = pd.DataFrame()
    
    for disasterno_temp_single in EMDAT_disasterno_temp:
        EMDAT_affected_shape = EMDAT_affected_shape.append(GDIS[GDIS.disasterno == disasterno_temp_single])

    EMDAT_temp_centroid = EMDAT_affected_shape['geometry'].unary_union.centroid
    EMDAT_points_matched.geometry.iloc[EMDAT_counter] = EMDAT_temp_centroid

EMDAT_points_matched = gpd.GeoDataFrame(EMDAT_points_matched,geometry='geometry')
EMDAT_points_matched.crs = crs
EMDAT_points_matched.to_file(path_world_map+'EMDAT_centroids_matched.shp')    
```
